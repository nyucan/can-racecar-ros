```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.load('1-01-01-01.npy')
data.shape

K = [1, 0.1, 0.1, 0.1]
err_d = data[:,0]
err_psi = data[:,1]
err_dot_d = data[:,2]
err_dot_psi = data[:,3]
steering = data[:,4]

fig, axs = plt.subplots(2, 2, figsize=(15, 8))
fig.suptitle('Error Metrics, K=[{:.1f}, {:.1f}, {:.1f}, {:.1f}]'.format(*K))
axs[0, 0].plot(err_d)
axs[0, 0].set_title('Distance')
axs[0, 1].plot(err_psi, 'tab:orange')
axs[0, 1].set_title('Psi')
axs[1, 0].plot(err_dot_d, 'tab:green')
axs[1, 0].set_title('Dot(Distance)')
axs[1, 1].plot(err_dot_psi, 'tab:red')
axs[1, 1].set_title('Dot(Psi)')
plt.show()

plt.figure(2, figsize=(15, 3))
plt.title('Steering')
plt.plot(steering)
plt.show()
```
